```python
# Import OpenCV and NumPy
import cv2
import numpy as np

# Read a color image
img_colour = cv2.imread(r"C:\Users\Administrator\Downloads\NED Gate.jpg", 1)  # 1 = color

# Read the same image in grayscale
img_grayscale = cv2.imread(r"C:\Users\Administrator\Downloads\NED Gate.jpg", 0)  # 0 = grayscale

# Use matplotlib to display the image
from matplotlib import pyplot as plt

# Convert from BGR (OpenCV default) to RGB (matplotlib expects RGB)
plt.imshow(cv2.cvtColor(img_colour, cv2.COLOR_BGR2RGB))
plt.axis("off")  # Hide axes
plt.show()
```

Resize

```python
# Resize the image to 150x150 pixels
img_resized = cv2.resize(img_colour, (150, 150))

# Optional: Resize using scaling factors instead
# img_resized = cv2.resize(img_colour, (0, 0), fx=1.5, fy=1.5)

# Display the resized image
plt.axis("off")
plt.imshow(cv2.cvtColor(img_resized, cv2.COLOR_BGR2RGB))
plt.show()
```

Pixel Manipulation

```python
# Pixel Manipulation
new_img = img_colour.copy()
new_img[70:125, 25:75] = 0

# Cropping
new_img2 = img_colour[85:115, 110:570]

plt.axis("off")
plt.imshow(cv2.cvtColor(new_img, cv2.COLOR_BGR2RGB))
plt.show()

plt.axis("off")
plt.imshow(cv2.cvtColor(new_img2, cv2.COLOR_BGR2RGB))
plt.show()
```

Translation

```python
# Get image dimensions
rows, cols, channels = img_colour.shape

# Define the translation matrix (shift x=180, y=60)
M = np.float32([[1, 0, 180], [0, 1, 60]])

# Apply translation
translated = cv2.warpAffine(img_colour, M, (cols, rows))

# Show the translated image
plt.axis("off")
plt.imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB))
plt.title("Translated Image")
plt.show()
```

Rotation

```python
# Rotation
# Rotate image from center of image with an angle of 45 degrees at the same scale.
rotation_matrix = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)  # 1st: Center Point, 2nd: Angle, 3rd: Scale
# Apply the transformation
rotated = cv2.warpAffine(img_colour, rotation_matrix, (cols, rows))

# Show the rotated image
plt.axis("off")
plt.imshow(cv2.cvtColor(rotated, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_grey = cv2.imread(r"C:\Users\Administrator\Downloads\Contour.jpeg", 0)
```

Edge Detection

```python
# Edge Detection
edges = cv2.Canny(img_grey, 30, 150)

# Display edges in the image
plt.imshow(edges, cmap='gray')
plt.axis("off")
plt.show()
```

Contours

```python
# Threshold the image
_ , thresholded_image = cv2.threshold(img_grey, 220, 255, cv2.THRESH_BINARY_INV)

# Detect contours
contours, _ = cv2.findContours(thresholded_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

# Draw the detected contours; -1 means draw all contours
cv2.drawContours(img_grey, contours, -1, (0, 255, 0), 3)

# Display the image with contours
plt.imshow(cv2.cvtColor(img_grey, cv2.COLOR_BGR2RGB))
plt.axis("off")  # Hide axes
plt.show()
```
